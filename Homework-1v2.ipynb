```python
import string
import random
random.seed(0)

class People:
    def __init__(self, my_first_names = " ", my_middle_names = " ", my_last_names = " ", order = "first_name_first"):
        self.first_names = my_first_names
        self.middle_names = my_middle_names
        self.last_names = my_last_names
        self.my_order = order
        
    
    def __iter__(self):
        return my_NameIterator(self)
    
    def __call__(self):
        for i in sorted(self.last_names):
            print(i)
        
        
class my_NameIterator:
    def __init__(self,name):
        self.name = name
        self.index = 0
        
    
    def __iter__(self):
        return (self)
    
    def __next__(self):
        if (self.index < len(self.name.first_names)) and (self.name.my_order == "first_name_first"):
            theFullName = self.name.first_names[self.index]+" "+self.name.middle_names[self.index]+" "+self.name.last_names[self.index]
            self.index+=1 
            return theFullName
        elif (self.index < len(self.name.first_names)) and (self.name.my_order == "last_name_first"):
            theFullName = self.name.last_names[self.index]+" "+self.name.first_names[self.index]+" "+self.name.middle_names[self.index]
            self.index+=1 
            return theFullName 
        elif (self.index < len(self.name.first_names)) and (self.name.my_order == "last_name_with_comma_first"):
            theFullName = self.name.last_names[self.index]+", "+self.name.first_names[self.index]+" "+self.name.middle_names[self.index]
            self.index+=1 
            return theFullName
        
        raise StopIteration
    
 
    
class PeopleWithMoney(People):
    def __init__(self,my_first_names,my_middle_names,my_last_names,order,wealth):
        self.first_names = my_first_names
        self.middle_names = my_middle_names
        self.last_names = my_last_names
        self.my_order = order
        self.wealth = wealth
        People.__init__(self,my_first_names,my_middle_names,my_last_names,order ="first_name_first")
        
    def __iter__(self):
        return PeopleWithMoneyIterator(self,my_NameIterator(self))
    
    def __call__(self):
        self.theNames = [a + " " + b + " "+ c  for a,b,c in zip(self.first_names,self.middle_names,self.last_names)]
        result = [q +" "+str(r) for q,r in (zip(self.theNames,self.wealth))]
        ascending_results = [r for q,r in sorted(zip((self.wealth),result))]
        for values in ascending_results:
            print (values)
        
    
class PeopleWithMoneyIterator:
    def __init__(self,money,value):
        self.money = money
        self.index = 0
        self.myValue = iter(value)
        
    def __iter__(self):
        return self
    
    def __next__(self):
        if (self.index < len(self.money.wealth)):
            myOutput = next(self.myValue)+" "+str(self.money.wealth[self.index])
            self.index += 1
            return myOutput
        raise StopIteration
        


first_names  = []
middle_names = []
last_names = []
#random.seed(0)

letters = string.ascii_lowercase


for item in range (10):
    myLetters = (''.join(random.choice(letters) for i in range(5)))
    myLetters2 = (''.join(random.choice(letters) for i in range(5)))
    myLetters3 = (''.join(random.choice(letters) for i in range(5)))
    first_names.append(myLetters)
    middle_names.append(myLetters2)
    last_names.append(myLetters3)

    
#print ("Firstname", first_names)
#print ("middleName", middle_names)
#print ("lastName", last_names)

print ("---------------------------------------------------------------------") 
print ("FIRST NAME FIRST STARTS HERE")
print ("---------------------------------------------------------------------")
nameList_1 = People(first_names,middle_names,last_names)

for item in nameList_1:
    print(item)

print ("---------------------------------------------------------------------") 
print ("LAST NAME FIRST STARTS HERE")
print ("---------------------------------------------------------------------")
nameList_2 = People(first_names,middle_names,last_names,order = "last_name_first")

for item2 in nameList_2:
    print(item2)
    

print ("---------------------------------------------------------------------") 
print ("LAST NAME FIRST AND COMMA STARTS HERE")
print ("---------------------------------------------------------------------")
nameList_3 = People(first_names,middle_names,last_names,order = "last_name_with_comma_first")

for item3 in nameList_3:
    print(item3)
    
print ("---------------------------------------------------------------------") 
print ("SORTED LAST NAME STARTS HERE")
print ("---------------------------------------------------------------------")
nameList_1()

print ("---------------------------------------------------------------------") 
print ("FIRSNAME MIDDLENAME LASTNAME WEALTH STARTS HERE")
print ("---------------------------------------------------------------------")
my_cash = []
for my_item in range(10):
    my_cash.append(random.choice(range(0,1000)))
moneyPeople = PeopleWithMoney(first_names,middle_names,last_names,order = "first_name_first",wealth = my_cash)
for item4 in moneyPeople:
    print(item4)

print ("---------------------------------------------------------------------") 
print ("SORTED FIRSNAME MIDDLENAME LASTNAME WEALTH STARTS HERE")
print ("---------------------------------------------------------------------")
moneyPeople()        
        
```
